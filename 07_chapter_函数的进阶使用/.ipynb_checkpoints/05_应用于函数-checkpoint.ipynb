# 数据容器中通用函数的进阶使用

编写一个程序，分析学生成绩列表，计算以下内容：

```python
students = [
    {"name": "小明", "score": 75},
    {"name": "小红", "score": 88},
    {"name": "小刚", "score": 59},
    {"name": "小丽", "score": 92},
]
```

1. 班级的平均分；
2. 分数最高的学生姓名；
3. 分数最低的学生姓名；
4. 所有不及格（分数小于60分）的学生姓名。
5. 按成绩排序

````python
students = [
    {"name": "小明", "score": 75},
    {"name": "小红", "score": 88},
    {"name": "小刚", "score": 59},
    {"name": "小丽", "score": 92},
]
````

````python
result = 0
for student in students:
    result += student['score']

print(result / len(students))
````

1.min(iterable, key=函数)

key=函数 作用是告诉 Python：比较时要以什么为标准。

````python
min(students, key=lambda x: x['score'])
````

````python
students = [
    {"name": "小明", "score": 75},
    {"name": "小红", "score": 88},
    {"name": "小刚", "score": 59},
    {"name": "小丽", "score": 92},
]
````

2.max(iterable, key=函数)

````python
max(students, key=lambda x: x['score'])
````

````python
def fun(x):
    return x['score']

for student in students:
    score = fun(student)
    print(score)
````

3.map(function, iterable)：对可迭代对象的每一个元素应用函数，并返回迭代器

````python
result = map(lambda x: x**2, [1, 2, 3, 4, 5, 6])
list(result)
````

````python
list_1 = ['hello world', 'hello suining', 'hello yiban']
list(map(lambda x:x.upper(), list_1))
````

4.filter(function, iterable)

````python
list_1 = [100, 98, 87, 59, 58, 48]
list(filter(lambda x: x>=60, list_1))
# [True, True, Ture, False, False, False]
````

````python
students = [
    {"name": "小明", "score": 75},
    {"name": "小红", "score": 88},
    {"name": "小刚", "score": 59},
    {"name": "小丽", "score": 92},
]

# 所有不及格（分数小于60分）的学生姓名。
list(filter(lambda x:x['score']<60, students))
````

5.sorted(iterable, key=..., reverse=...)

````python
students = [
    {"name": "小明", "score": 75},
    {"name": "小红", "score": 88},
    {"name": "小刚", "score": 59},
    {"name": "小丽", "score": 92},
]
````

````python
sorted(students, key=lambda x: x['score'], reverse=False)
````

6.all()：判断一个集合中的所有元素是否都为真（非0、非空、非False）

7.any()：判断一个集合中的有一个元素是否都为真（非0、非空、非False）
