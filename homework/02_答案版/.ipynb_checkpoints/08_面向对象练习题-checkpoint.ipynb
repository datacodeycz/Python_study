请定义一个 Rectangle 类，满足以下要求：

1.将`宽（width）`和`高（height）`设置为私有属性（即以 __width 和 __height 命名）

2.使用构造函数 `__init__()` 初始化宽和高

3.使用`@property`和 `@setter`方法，分别为宽和高提供读取和写入接口，要求宽和高必须为正数，否则设置时给出提示（例如“宽度必须为正数”）

4.定义`area()`方法，返回矩形的面积

5.定义`is_square()`方法，判断该矩形是否为正方形（宽==高）

```python
class Rectangle:
    def __init__(self, width, height):
        self.__width = None
        self.__height = None
        self.width = width  # 使用 setter 方法设置宽度
        self.height = height  # 使用 setter 方法设置高度

    @property
    def width(self):
        return self.__width

    @width.setter
    def width(self, value):
        if value <= 0:
            raise ValueError("宽度必须为正数")
        self.__width = value

    @property
    def height(self):
        return self.__height

    @height.setter
    def height(self, value):
        if value <= 0:
            raise ValueError("高度必须为正数")
        self.__height = value

    def area(self):
        return self.__width * self.__height

    def is_square(self):
        return self.__width == self.__height
# 测试代码
if __name__ == "__main__":
    rect = Rectangle(5, 10)
    print(f"宽度: {rect.width}, 高度: {rect.height}, 面积: {rect.area()}")
    print(f"是否为正方形: {rect.is_square()}")

    rect.width = 10
    print(f"修改后宽度: {rect.width}, 高度: {rect.height}, 面积: {rect.area()}")
```

使用Python程序设计，编写银行账户类

要求：\
1.定义类 `BankAccount`，包含`账号（account）`、`姓名（name）`、`余额（balance）`

2.使用构造方法初始化属性

3.添加 `deposit(amount)` 方法：实现存款功能

4.添加 `withdraw(amount) `方法：余额足够则扣款，否则提示“余额不足”

5.添加 `get_balance()` 方法：返回当前余额

```python
class BankAccount:
    def __init__(self, account, name, balance=0):
        self.account = account
        self.name = name
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"存款成功，当前余额为：{self.balance}")
        else:
            print("存款金额必须大于0")

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print(f"取款成功，当前余额为：{self.balance}")
        else:
            print("余额不足")

    def get_balance(self):
        return self.balance
# 测试代码
if __name__ == "__main__":
    account = BankAccount("123456789", "张三", 1000)
    print(f"账号：{account.account}, 姓名：{account.name}, 初始余额：{account.get_balance()}")
    account.deposit(500)
    account.withdraw(300)
    account.withdraw(1500)  # 测试余额不足
    print(f"最终余额：{account.get_balance()}")
```

使用Python程序设计，编写动物类（涉及继承）

要求:\
1.创建一个 Animal 父类，有属性 name 和方法 speak()（输出“动物在叫”）

2.创建一个 Dog 子类，继承自 Animal，重写 speak() 方法，输出“狗叫：汪汪汪”

3.创建一个 Cat 子类，继承自 Animal，重写 speak() 方法，输出“猫叫：喵喵喵”

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        print("动物在叫")
class Dog(Animal):
    def speak(self):
        print("狗叫：汪汪汪")
class Cat(Animal):
    def speak(self):
        print("猫叫：喵喵喵")
# 测试代码
if __name__ == "__main__":
    animal = Animal("动物")
    animal.speak()

    dog = Dog("狗")
    dog.speak()

    cat = Cat("猫")
    cat.speak()
```
