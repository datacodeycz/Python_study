3.文件`numbers.txt`中存储了一系列整数，每行一个。请读取所有整数并输出它们的和。

````python
with open('./data/number.txt', 'r', encoding='utf-8') as f:
    numbers = f.readlines()

result = 0
for num in numbers:
    if '\n' in num:
        num = num[:-1]
        # print(num)
    result = result + int(num)
result
````

7.文件`poem.txt`中包含一段英文诗句，请读取该文件，并统计其中每个单词出现的次数，结果按单词首字母排序后输出。

````python
with open('./data/poem.txt', 'r', encoding='utf-8') as f:
    text = f.read()

# 去除标点符号（简单处理逗号和句号）
for ch in ",.":
    text = text.replace(ch, "")

text
````

````python
# 将文本拆分为单词
words = text.lower().split()
words
````

````python
# 统计词频
word_count = {}
for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1
````

````python
# 按单词首字母排序输出
for word in sorted(word_count.keys(), key=lambda x: x[0]):
    print(f"{word}: {word_count[word]}")
````

8.编写程序将九九乘法表写入文件`multiplication.txt`，每一行格式如下：
```
1 * 1 = 1   
1 * 2 = 2   2 * 2 = 4   
1 * 3 = 3   2 * 3 = 6   3 * 3 = 9   
1 * 4 = 4   2 * 4 = 8   3 * 4 = 12   4 * 4 = 16   
1 * 5 = 5   2 * 5 = 10   3 * 5 = 15   4 * 5 = 20   5 * 5 = 25   
1 * 6 = 6   2 * 6 = 12   3 * 6 = 18   4 * 6 = 24   5 * 6 = 30   6 * 6 = 36   
1 * 7 = 7   2 * 7 = 14   3 * 7 = 21   4 * 7 = 28   5 * 7 = 35   6 * 7 = 42   7 * 7 = 49   
1 * 8 = 8   2 * 8 = 16   3 * 8 = 24   4 * 8 = 32   5 * 8 = 40   6 * 8 = 48   7 * 8 = 56   8 * 8 = 64   
1 * 9 = 9   2 * 9 = 18   3 * 9 = 27   4 * 9 = 36   5 * 9 = 45   6 * 9 = 54   7 * 9 = 63   8 * 9 = 72   9 * 9 = 81
```

````python
with open('./data/multiplication.txt', mode='w', encoding='utf-8') as f:
    for y in range(1, 9 + 1):
        for x in range(1, y+1):
            f.write(f'{x} * {y} = {x * y}  ')
        f.write('\n')
````
