6.创建一个函数，实现将前n项斐波那契数列存储到列表中,并返回

```python
def faio(n):
    a = 0
    b = 1
    result = []
    if n == 0:
        # print(a)
        result.append(a)
    if n == 1:
        # print(b)
        result.append(b)
    for _ in range(n):
        a, b = b, a + b
        # print(a)
        result.append(a)
    return result

faio(10)
```

8.给定学生信息字符串列表：["20230101-张三", "20230102-李四", "20230103-王五"]，请将其拆分为学号和姓名两个列表，分别输出。

```python
lst = ["20230101-张三", "20230102-李四", "20230103-王五"]
ids = []
names = []
for str_infor in lst:
    id_, name = str_infor.split('-')
    ids.append(id_)
    names.append(name)

ids, names
```
