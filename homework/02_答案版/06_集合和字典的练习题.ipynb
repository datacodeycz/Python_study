20.使用嵌套字典表示多个学生的选课信息。

每个学生包含：

`name`（姓名）
`courses`（课程信息，是一个嵌套字典）

    `python程序设计`
    `mysql库数据`
    `network`
    
    每门课程包含：
        credit（学分）
        teacher（任课老师）
        required（是否必修）

要求：

1.添加 2 位学生，每人选修至少 2 门课程

2.打印每个学生的所有课程信息

3.统计并输出每位学生的总学分

````python
students = {
    "Alice": {
        "courses": {
            "python程序设计": {
                "credit": 3,
                "teacher": "张老师",
                "required": True
            },
            "mysql库数据": {
                "credit": 2,
                "teacher": "李老师",
                "required": False
            }
        }
    },
    "Bob": {
        "courses": {
            "network": {
                "credit": 4,
                "teacher": "王老师",
                "required": True
            },
            "python程序设计": {
                "credit": 3,
                "teacher": "张老师",
                "required": True
            }
        }
    }
}
# 打印每个学生的所有课程信息
for student_name, student_info in students.items():
    print(f"学生: {student_name}")
    for course_name, course_info in student_info["courses"].items():
        print(f"  课程: {course_name}")
        print(f"    学分: {course_info['credit']}")
        print(f"    任课老师: {course_info['teacher']}")
        print(f"    是否必修: {'是' if course_info['required'] else '否'}")
    print()
# 统计并输出每位学生的总学分
for student_name, student_info in students.items():
    total_credit = sum(course["credit"] for course in student_info["courses"].values())
    print(f"学生: {student_name} 的总学分: {total_credit}")
````
