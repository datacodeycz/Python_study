## 第二章和第三章编程题作业

13. 给定整数 156，使用内置函数将其转换为二进制、八进制、十六进制的字符串表示。

`bin`:将整数（十进制）类型的转换为（二进制）字符串\
`oct`:将整数（十进制）类型的转换为（八进制）字符串\
`hex`:将整数（十进制）类型的转换为（十六进制）字符串

14. 给定字符串 "1010110"(表示二进制|数)，先用 int(..., 2) 将其转换为十进制整数，再分别转换为八进制和十六进制的字符串。

15. 给定字符串 "0o237"(表示八进制数)，先用 int(..., 8) 转换为整数，再转换为二进制和十六进制的字符串。

16. 给定字符串 "0x1A3"(表示十六进制数)，先用 int(..., 16) 转换为整数，再转换为二进制和八进制字符串。

17. 将浮点数 123.45 通过 int()转换为整数，再将结果分别转换为二进制、八进制、十六进制的字符串。

18. 将布尔值 True 分别转换为整数、浮点数和字符串。

19. 将布尔值 False 分别转换为整数、浮点数和字符串。

20. 给定整数 1023，先用 bin()、oct()、hex() 得到其二进制、八进制、十六进制字符串，再利用 int()将这些字符串转换回整数，验证结果均为 1023。

21. 给定字符串 "1111111111"(无前缀的二进制数)，用 int(..., 2) 转换为整数，再转换为浮点数。

22. 给定字符串 "3.14159"，先转换为浮点数，再用 int() 转换为整数。

23. 给定整数 256，使用内置函数获取其二进制、八进制、十六进制的字符串表示，然后分别利用 int(…, base) 转换回整数。

24. 将浮点数 0.0 分别转换为整数和布尔值。

25. 在 Python 中计算表达式 True + False，并说明其数据类型和结果。

True 表示 1
False 表示 0

在进行算术运算时，会转换为数字类型进行算术运算，所以得到的结果也是数字类型的

26. 给定二进制字符串 "0b101101"，先转换为整数，再加上 10，输出最终结果。

27. 给定八进制字符串 "0o764"，转换为整数后乘以 2，输出结果。

28. 给定十六进制字符串 "0xFF"，转换为整数后减去 100，输出结果。

29. 给定整数 -45，使用 bin() 转换为二进制字符串，并说明结果的含义。

30. 给定浮点数 100.99，先用 int() 转换整数，再获取其二进制、八进制、十六进制的字符串表示。

31. 给定字符串 "False"，直接用 bool() 转换成布尔值，并说明结果。

字符串转换为布尔类型的时候，非空表示True  空表示False
