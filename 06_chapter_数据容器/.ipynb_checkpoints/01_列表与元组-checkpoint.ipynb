## 数据容器-list(列表)
列表是什么：列表是一种数据容器

列表的特点：1.有序；2.可以存储重复的元素；3.可以存储不同类型的元素；4.可以被修改

### 列表的创建

````python
# 2. 使用list()创建空列表 
list_2 = list()  # 只有数据容器才能够转换为其他数据容器
print(list_2)
print(type(list_2))
````

#### 列表的特点4：可以被修改
意味着：可以被增删改查

### 列表的操作

1.访问列表的中的元素\
```列表名[num]``` \
注意：表中存储的每一个元素都有对应的索引，我们可以通过索引去访问对应的元素 索引是 开始到末尾是从0开始到length-1，而末尾到开始是从-1到-length\
2.修改列表中对应的元素-- 通过下标去修改列表中对应的元素\
 ``列表名[num] = 需要修改的值``\
3.新增元素<br>
(1)```列表名.insert(参数1，参数2)``` 在指定的下标位置，插入指定的元素\
(2)```列表名.append(元素)``` 将指定元素，增加到列表的尾部\
(3)```列表名.extend(容器)```将参数容器的内容依次取出，插入到列表的尾部\
4.删除元素<br>
(1)``del 列表名[下标]``\
(2)``列表名.pop(下标)`` 返回被删掉的元素\
(3)```列表.remove(元素)``` 从正方向，删除出现的第一个元素\
5.清空元素<\
```列表名.clear()``` 清空列表\
6.统计元素个数\
```len(列表名)```\
7.通过元素值查找元素对应的下标--注意返回的是正下标\
```列表名.index(元素值)```\
8.统计某个元素在列表中出现的次数\
```列表名.count```

````python
list_1 = ['a', 'b', 'c', 'd', 'e', 'f']
# print(list_1.pop(2))
# # print(list_1)
print(list_1.extend([1, 2, 3]))
````

````python
# 1.访问列表的中的元素
list_1 = ['a', 'b', 'c', 'd', 'e', 'f']
print(list_1[3])
````

````python
print(list_1[-3])
````

````python
# 2.通过下标修改列表中对应的元素
list_1 = ['a', 'b', 'c', 'd', 'e', 'f']
list_1[2] = 'z'
print(list_1)
````

````python
list_1 = ['a', 'b', 'c', 'd', 'e', 'f']
list_1[-3] = 'z'
print(list_1)
````

````python
# 3. 增加元素
# 3.1 在指定的下标位置增加元素
list_1 = ['a', 'b', 'c', 'd', 'e', 'f']
print(list_1.insert(2, 'z'))
print(list_1)
````

````python
# 3.2在列表末尾增加元素
list_1 = ['a', 'b', 'c', 'd', 'e', 'f']
print(list_1.append('g'))
print(list_1)
````

````python
# 3.3 在列表末尾增加一组元素
list_1 = ['a', 'b', 'c', 'd', 'e', 'f']
print(list_1.extend([1, 2, 3, 4]))  # 这里只能填写一个参数，并且这个参数必须是一个数据容器
print(list_1)
````

````python
# 4. 删除元素
# 4.1使用del 删除指定下标元素
list_1 = ['a', 'b', 'c', 'd', 'e', 'f']
del list_1[2]
print(list_1)
````

````python
# 4.2 使用pop()删除列表中的指定下标的元素
# 不填参数，默认删除最后一个
list_1 = ['a', 'b', 'c', 'd', 'e', 'f']
list_1.pop()
print(list_1)
````

````python
list_1 = ['a', 'b', 'c', 'd', 'e', 'f']
print(list_1.pop())
print(list_1)
````

````python
list_1 = ['a', 'b', 'c', 'd', 'e', 'f']
list_1.pop(2)
print(list_1)
````

````python
# 4.3 使用remove删除列表中第一个匹配到的元素
list_1 = ['a', 'b', 'c', 'b', 'b', 'b']

print(list_1.remove('b'))
print(list_1)
````

````python
# 5. 清空列表
list_1 = ['a', 'b', 'c', 'b', 'b', 'b']
print(list_1.clear())
print(list_1)
````

````python
# 6.统计元素的个数
list_1 = ['a', 'b', 'c', 'd', 'e', 'f']
n = len(list_1)
print(n)
````

````python
# 7.通过元素值查找元素对应的下标(遇到的第一个)
list_1 = ['a', 'b', 'c', 'b', 'b', 'b']
print(list_1.index('b'))  # 返回正下标
````

````python
# 8.统计某个元素在列表中出现的次数
list_1 = ['a', 'b', 'c', 'b', 'b', 'b']
print(list_1.count('b'))
````

### 列表的遍历
1. 使用while循环，遍历列表中的元素
2. 使用for循环，遍历列表中的元素

````python
# 使用while循环遍历
list_1 = ['a', 'b', 'c', 'd', 'e', 'f']
n = len(list_1)
i = 0 
while i<=n-1:
    print(list_1[i])
    i = i + 1
````

````python
list_1 = ['a', 'b', 'c', 'd', 'e', 'f']
n = len(list_1)
i = -1
while i >= -n:
    print(list_1[i])
    i = i - 1
````

````python
# 使用for循环遍历
list_1 = ['a', 'b', 'c', 'b', 'b', 'b']

n = 0
for i in list_1:
    if i == 'b':
        print(n)
        break
    n = n + 1
````

小结：
1. 使用while循环可以人为的控制条件并输出的结果，例如 我可以控制输出下标为奇数对应的元素，只需要将 index = 1  index = index + 2即可
2. 使用for循环，更简单，更直观的体现了遍历的过程，并且只能将所有元素进行遍历，但是这不妨碍for循环控制输出的结果，可以借助if语句和continue关键字

### 列表小结
通过学习，我们可以总结列表的一些特点，到最后我们学完所有容器，再做一次总结\
1.列表可以容纳多个元素，上限为2^63 - 1\
2.列表可以容纳不同数据类型的元素（跟Java作比较）\
3.列表中的数据都是有下标，按照下标的顺序排列\
4.列表中的数据是可以重复出现的\
5.列表中的数据是可以被修改的

## 数据容器-tuple(元组)
有时候我们需要一个容器来存放重要的数据，这个重要的数据是不允许被修改的，如果我们还用列表的话，列表中的元素是会被修改的，所以我们要引入一个新的数据容器-元组，**元组中的元素一旦定义确定后，是不可被修改的**

元组的特点：1.有序；2.可以存储重复的元素；3.可以存储不同类型的元素；4.不可被修改

### 元组的创建
定义元组使用小括号，并且使用逗号隔开各个数据，数据**可以是不同数据类型**

````python
# 元组的创建
# tuple
# tuple
# 1.创建空元组方式1：使用()
tuple_1 = ()
print(tuple_1)
print(type(tuple_1))
````

````python
# 2.创建空的元组方式2：tuple()
tuple_2 = tuple()
print(tuple_2)
print(type(tuple_2))
````

````python
# 3.使用()或tuple()创建有元素的元组
tuple_3 = (1, 2, 3, 4,)
print(tuple_3)
````

````python
tuple_4 = tuple([1, 2, 3])
print(tuple_4)
````

````python
tuple_5 = (1,)
print(tuple_5)
print(type(tuple_5))
````

注意：当只有一个元素的时候，在后面加一个逗号

````python
# 4.两个元组拼接成一个元组
tuple_6 = ('a', 'b', 'c')
tuple_7 = (1, 2, 3)
tuple_8 = tuple_6 + tuple_7
print(tuple_8)
````

````python
list_1 = [1, 2, 3]
tuple_1 = ('a', 'b', 'c', list_1)
print(tuple_1)

list_1.append(4)
print(tuple_1)
````

#### 元组的特点1：有序

#### 元组的特点2：可以存储重复的元素

````python
tuple_1 = ('a', 'b', 'a', 'a', 'e')
print(tuple_1)
````

#### 元组的特点3：可以存储不同类型的元素

````python
tuple_1 = (1, 1.0, True, 1+2j, '123', [1, 2, 3])
print(tuple_1)
````

#### 元组的特点4：不可被修改
不能进行增删改的操作，只能进行查找的操作

### 元组的操作
由于元组在创建的时候就已经确定的元组中的元素，所以无法进行增删改的操作，只能进行查找操作<br>
操作有<br>
1.元组名\[num\]:访问元组中对应下标的元素值<br>
2.index():查找某个元素，返回对应的下标，如果不存在会报错<br>
3.count()：统计某个元素在当前元组中出现的次数<br>
4.len(元组):统计元组内的元素个数<br>

````python
# 1.访问元素值
tuple_1 = ('a', 'b', 'c', 'd', 'e')
print(tuple_1[-1])
````

````python
# 2.index()
tuple_1 = ('a', 'b', 'c', 'd', 'e')
print(tuple_1.index('c'))
````

````python
# 3.count()
tuple_1 = ('a', 'b', 'a', 'a', 'e')
print(tuple_1.count('a'))
````

````python
# 4.len()
len(tuple_1)
````

## 元组的遍历
while循环语句<br>
for循环语句<br>

````python
# while循环语句
# 正下标循环
tuple_1 = ('a', 'b', 'c', 'd', 'e', 'f')
i = 0
n = len(tuple_1)
while i <= n - 1:
    print(tuple_1[i])
    i = i + 1
````

````python
# 负下标
tuple_1 = ('a', 'b', 'c', 'd', 'e', 'f')
i = -1
n = len(tuple_1)
while i >= -n:
    print(tuple_1[i])
    i = i - 1
````

````python
# for循环遍历
tuple_1 = ('a', 'b', 'c', 'd', 'e', 'f')

for i in tuple_1:
    print(i)
````

### 元组小结
1.可以存储多个数据<br>
2.可以存放不同数据类型的数据<br>
3.元组中的数据都是有对应的下标的<br>
4.允许出现重复的数据<br>
5.**不可以被修改**<br>
6.可以使用for循环遍历查找<br>
