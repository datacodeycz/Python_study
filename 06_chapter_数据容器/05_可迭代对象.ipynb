# 可迭代对象
## 可迭代对象的概念
可迭代对象是指：可以使用 for 循环进行遍历的对象。\
只要一个对象实现了 `__iter__()` 方法（或说支持迭代协议），就叫做“可迭代对象”。

```python
list_1 = [1,2,3,4,5]

list_1.__iter__()  # list有这个方法
```

## 可迭代对象的类别

常见的可迭代对象包括：\
1.列表；2.元组；3.字符串；4.集合；5.字典；\
6.range；6.视图对象 `dict.keys()`, `dict.values()`,`dict.items()`  7.文件对象 `open(file)`


## 可迭代对象的作用
为什么要有可迭代对象：
1. 统一的遍历接口，都可以用for循环迭代实现
2. 节省内存，支持惰性计算（数据不会一次性全部读取到内存中）
3. 支持通用的工具函数和语法糖\
   1.for循环迭代遍历\
   2.`in`操作\
   3.解包\
   4.`sum()` `max()` `min()` 等操作
