# random模块的学习
random模块主要是用于生成随机数，通过不同的需求生成不同的随机数\
如果要使用random模块首先需要要导入random模块

```python
# 导入random模块
import random
```

## 基本随机数生成
- **`random.random()`**：生成一个 `[0.0, 1.0)` 范围内的随机浮点数。
- **`random.uniform(a, b)`**：生成一个 `[a, b]` 范围内的随机浮点数。

```python
# 1.random()方法
'''
功能：返回一个随机浮点数
输入（参数）：无
    
输出（返回值）：
    返回随机生成的一个实数，范围在[0,1)内
'''
# 使用random()方法
random.random()
```

```python
# 从范围内生成随机数
```

```python
# 练习
# 我们现在需要生成一个1-10的随机数 0-1 * 9 + 1
print(random.random() * 9 + 1)
```

```python
#  我们现在需要生成一个x-y的随机数
x = 2
y = 10

random.random() * (y - x) + 2
```

## 整数随机生成
- **`random.randint(a, b)`**：生成一个 `[a, b]` 范围内的随机整数。
- **`random.randrange(start, stop[, step])`**：从 `range(start, stop, step)` 中随机返回一个整数。

```python
# 2.randint(a, b)
'''
功能：返回随机整数，整数的范围有参数决定
输入（参数）
    1.随机整数的左区间值
    2.随机整数的右区间值
    例如：randint(a,b) 表示范围为 [a,b]
输出（返回值）
    返回 由参数决定的区域之间 随机生成的一个整数
'''

random.randint(2, 100)
```

```python
# 从范围内生成随机数 random.randrange(start, end, step)
random.randrange(1, 10, 2)
```

## 随机选择与打乱
- **`random.choice(seq)`**：从序列 `seq` 中随机选择一个元素。
- **`random.choices(population, weights=None, k=1)`**：从 `population` 中随机选择 `k` 个元素，可以指定权重。
- **`random.sample(population, k)`**：从 `population` 中随机抽取 `k` 个不重复元素。
- **`random.shuffle(seq)`**：将序列 `seq` 中的元素随机打乱。

```python
# 随机选择 random.choice()
list_1 = ['a', 'b', 'c', 'd', 'e', 'f']
random.choice(list_1)  # 
```

```python
# 随机选择多个 random.choices()
random.choices(list_1, weights=[1, 2, 3, 0, 0, 0]  , k=4)  # 返回的一个列表
```

```python
# 随机抽取不重复的 random.sample()
list_1 = ['a', 'b', 'c', 'd', 'e', 'f']
random.sample(list_1, 4)
```

```python
# 打乱顺序 random.shuffle
list_1 = ['a', 'b', 'c', 'd', 'e', 'f']
result = random.shuffle(list_1)  # 改变了原来的列表
result, list_1
```

## 随机分布

- **`random.gauss(mu, sigma)`**：生成一个符合正态分布的随机数。
- **`random.expovariate(lambd)`**：生成一个符合指数分布的随机数。
- **`random.uniform(a, b)`**：生成均匀分布的随机数（之前提到）。

$$
\sigma
\lambda
$$


```python
# 正态分布  random.gauss()
random.gauss(85, 1)
```

```python
# 指数分布 random.expovariate()
random.expovariate(2)
```

```python
# 生成均匀分布的随机数
random.uniform(1,10)
```

## 随机种子
`random.seed(a=None)`：初始化随机数生成器的种子, a 是种子值。
- **用途**：使随机结果可重复，方便调试。

时间戳：1970

```python
import time
time.time()
```

```python
def random_slef(time):
    return time *12709871204  / 213871 / 136821083120 / 12371270 

random_slef(time.time())
```

```python
# 使用种子
random.seed(100)

random.random()
```

```python
random.random()
```

```python
random.seed(100)
random.random()
```

## 种子的概念
在python的内部有一个随机种子，它决定了生成的随机数是什么，当我们没有人为去改变我们的种子时，随机种子会根据当前的系统时间做为默认值，而当前系统的时间正常情况下，是不一样的，所以我们获取到的随机数是不一样的，但是有时候我们需要使用相等的随机数做测试，所以我们需要指定随机种子的值，这样就可以保证该随机种子下产生的随机数是一样的，测试结果与预期结果也能一致
