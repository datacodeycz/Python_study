# 面向对象

面向对象是一种强大的编程范式，旨在通过类和对象的概念来组织和封装代码。本章将带你学习面向对象编程的核心思想和应用场景，内容包括：
1. **基本概述**：理解类和对象的定义以及它们在编程中的作用。
2. **成员变量和方法**：掌握如何定义和使用类中的属性和方法。
3. **类属性、类方法和静态方法**：学习不同类型的方法及其使用场景。
4. **魔术方法**：深入了解 Python 提供的特殊方法（如 `__init__` 和 `__str__`）。
5. **封装与私有化**：理解如何保护数据及实现信息隐藏。
6. **继承与扩展**：学习如何通过继承实现代码复用及功能扩展。
7. **多态与灵活性**：掌握多态的概念及其在提高代码灵活性方面的作用。

成员变量：对象属性   `对象.属性 = 属性值`   成员变量

成员方法：对象的行为 类里面写的方法，第一个参数为self的函数 叫做成员方法

类属性：在类里面写的变量 ，就是类属性  

类方法：在类里面写的方法，上面有一个`@classmethod` 第一参数为cls 指向这个类对象

静态方法：在类里面写的方法，上面有一个`@staticmethod` （跟普通的函数一样，只是进一步的封装）


对象都可以使用这些方法

类对象 类属性和类方法和静态方法

类属性 定义的一些 常量
类方法 定义的一些 创建指定的对象

静态方法：主要是提供给类中的类方法和成员方法使用


面向对象三大特征-----封装、继承、多态\
封装-----安全性\
继承-----扩展性\
多态-----灵活性


封装：\
将成员变量和成员方法私有化\
私有的成员变量和私有的成员方法\
`__xxx`
通过公开的方法，去指定访问特有的私有成员变量

继承：
父类   子类， 子类继承父类
1. 子类能够使用父类的所有东西，但是父类不能使用子类的任何东西
2. 重写：父类有这个方法，但子类对这个方法进行改写，（函数名不变）
3. 扩展：父类没有这个方法，但子类需要这个方法，自己创建这个方法（重新创建一个函数）
4. 多继承（一般情况下，不要使用多继承）：一个子类，多个父类，这个子类拥有所有父类的内容，谁先继承，先用谁的

多态：
多个子类重写父类方法
