# 基本数据类型
常用的数据类型<br>
数字（number）
    **1.整数（int）
    2.浮点数（float）**
    3.复数（complex）
    4.布尔（bool）<br>
**字符串（str）：描述文本的一种数据类型，由字符组成<br>
列表（list）：有序的可变序列<br>
元组（tuple）：有序的不可变序列<br>
集合（set）：无序不重复集合<br>
字典（dict）：无序key—value集合**<br>

数据类型时指字面量的数据类型，不同表示的字面量有不同的数据类型，这样的目的是为了区分，不同类型的特点

```python
# int类型
print(1, 2)
# float类型
print(1.0, 2.2)
# complex类型
print(1+2j)  # 一个实部 一个虚部 
# bool类型
print(True, False)  # True 表示逻辑运算中的真， False 表示假

# 字符串类型
print("123")  # 字符串，就是字符组成的就是字符串， 数字、字母、符号这些东西都可以转换为字符，都可以用字符表示,这些都是字符串类型的数
```

注意：数字类型有几种表现形式：二进制，八进制，十进制（默认），十六进制

```python
# 创建一个二进制的整数
int_bin = 0b101
# 创建一个八进制的整数
int_oct = 0o777
# 创建一个十六进制的整数
int_hex = 0xabc
print(int_bin)
print(int_oct)
print(int_hex)
```

注意：我们这里我们只是输入的表现形式为其他进制，但是存储到程序中的还是十进制，如果我们要输出二进制，八进制，十六进制要使用对应的方法\
`bin`函数：传入数字类型，将数字类型转换为二进制数据，以`字符串`的形式返回\
`oct`函数：传入数字类型，将数字类型转换为八进制数据，以`字符串`的形式返回\
`hex`函数：传入数字类型，将数字类型转换为十六进制数据，以`字符串`的形式返回

```python
number_int = 28
# 转换为二进制输出结果
print(bin(number_int))
# 转换为八进制输出结果
print(oct(number_int))
# 转换为十六进制输出结果
print(hex(number_int))
```

浮点数也有两种表现形式：小数的形式，科学计数法的形式

```python
# 创建一个科学计数法的形式
print(1e-2)
```

注意：有可能我们使用科学计数法的形式导致小数点后面不够输出，会直接输出科学计数法的格式

```python
print(1e-17)
```

## 变量和数据类型的关系
变量是一个未知的，分配好空间的一个量\
数据类型是指字面量的数据类型，\
当我们使用赋值操作 ` a = 字面量`时，是将字面量存储的地址赋给了a，a执行这个变量的地址，a本身没有数据类型，当我们对a做一些非赋值的操作时，实际是操作的字面量

总结来说：变量就是取的名字，字面量就是实体

```python
# type()函数
print(type(1))
print(type('1'))
```

## 数据类型转换

在python中，我们经常会使用到的操作就是：数据类型的转换，python给我们提供了对应的函数转换为对应的数据类型，该函数的名字为数据类型对应的英文\
例如

    要转换为int类型 使用int函数
    要转换为float类型 使用float函数
    要转换为bool类型 使用bool函数

函数：就是可以实现一个具体功能的工具，仅此而已

```python
# int 转换为 float
number_int = 10
print(float(number_int))
# float 转换为 int
number_float = 1.2
print(int(number_float))
```

```python
# float转换为bool
number_float = 10.2
print(bool(number_float))
# bool转换为float
number_bool = True
print(float(number_bool))
```

```python
# int 转换为 bool
number_int = 1
print(bool(number_int))
# bool 转换为 int
number_bool = False
print(int(number_bool))
```

通过上述的代码我们可以得出一个结论: 
1. `数字类型`转换为`bool类型`时，**非0为True**  **0为False**
2. `bool类型`转换为`整数（浮点数）` **True为1（1.0）**  **False为0（0.0）**

```python
# int 转换为 complex
number_int = 28
print(complex(number_int))
# complex 转换为 int 不能实现
```

**其他数据类型转换为字符串类型**\
注意：其它的数据类型转换为字符串时，原始值不变，只是将数据类型转换成了字符串类型的

```python
# int 转换为 str
number_int = 1000
number_int_str = str(number_int)
print(number_int_str)
print(type(number_int_str))
```

```python
number_int = 0b111
number_int_str = str(number_int)
print(number_int_str)
print(type(number_int_str))
```

```python
# float转换为str
number_float = 1.99
number_float_str = str(number_float)
print(number_float_str)
print(type(number_float_str))
```

```python
number_float = 1e-17
number_float_str = str(number_float)
print(number_float_str)
print(type(number_float_str))
```

```python
# bool转换为str
bool_1 = False
bool_1_str = str(bool_1)
print(bool_1_str)
print(type(bool_1_str))
```

```python
# complex 转换为str
complex_1 = 1 + 2j
complex_1_str = str(complex_1)
print(complex_1_str)
print(type(complex_1_str))
```

```python
print(complex_1)
```

注意：这里复数的内容是(1+2j),是包括小括号的，转换为字符串，小括号也是在里面的

**字符串类型转换为其他数据类型**\
注意：字符串转换为其他数据类型时，字符串中的格式必须对应数据类型的格式

int类型表示：  1 2 3 4 5 \
float类型表示：  1.2  1e-1\
bool类型表示： True False\
complex类型表示：  1 + 2j

```python
# str 转换为 int  默认识别十进制，
str_1 = '100'
str_1_int = int(str_1)
print(str_1_int)
print(type(str_1_int))
```

```python
str_1 = '100.0'
# str_1_int = int(str_1)  # 不符合int类型
print(str_1_int)
print(type(str_1_int))
```

```python
str_1 = '0b1111'
str_1_int = int(str_1, 2)  # 让int这个函数能够识别为字符串中的二进制
print(str_1_int)
print(type(str_1_int))
```

```python
str_1 = '1111'
str_1_int = int(str_1, 2)  # 让int这个函数能够识别为字符串中的二进制
print(str_1_int)
print(type(str_1_int))
```

```python
str_1 = '0o777'
str_1_int = int(str_1, 8)  # 让int这个函数能够识别为字符串中的二进制
print(str_1_int)
print(type(str_1_int))
```

```python
str_1 = '0xfff'
str_1_int = int(str_1, 16)  # 让int这个函数能够识别为字符串中的二进制
print(str_1_int)
print(type(str_1_int))
```

```python
# str 转换为 float  它支持科学计数法的形式
str_1 = '1.99'
str_1_float = float(str_1)
print(str_1_float)
print(type(str_1_float))
```

```python
str_1 = '1e-17'
str_1_float = float(str_1)
print(str_1_float)
print(type(str_1_float))
```

```python
# str 转换为 bool  # 支持小写
str_1  = ' '
str_1_bool = bool(str_1)
print(str_1_bool)
print(type(str_1_bool))
```

```python
# str 转换为 complex  有无小括号都可以
str_1 = '(1+2j)'
str_1_complex = complex(str_1)
print(str_1_complex)
print(type(str_1_complex))
```

**总结**：
1. 字符串转换为其它数据类型的时候，字符串中的实数的其它进制的表现形式无法识别  可以通过int(……, 2)、int(……, 8)、int(……, 16)
2. float函数可以识别字符串中的科学计数法的形式
3. 字符转换为复数时，有()没()都可以
4. bool函数能识别到字符串中小写的True和False  非空字符串对应bool中的True  空字符串对应bool中的False
