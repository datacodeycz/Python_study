PEP 8（Python Enhancement Proposal 8）这是 Python 编程语言的官方编码风格指南。PEP 8 为 Python 代码的编写和格式提供了规范，以提高代码的可读性和一致性。
PEP 8 涵盖的内容包括：
- 缩进：建议使用 4 个空格进行缩进，而不是使用制表符（Tab）。
- 行长：每行代码的长度不应超过 79 个字符。
- 空行：模块级函数和类之间应有两个空行；方法或函数内部的代码块之间使用一个空行。
- 命名约定：建议使用小写字母和下划线的组合（如 variable_name）用于函数和变量，类名使用大写字母开头的驼峰命名法（如 ClassName）。
- 导入顺序：导入应分为三部分：标准库导入、第三方库导入、自己项目的导入，并且每部分之间应有空行分隔。

## 标识符

总结：标识符命名中，只允许出现：英文，中文，数字，下划线_<br>
ps：1.不推荐使用中文;2.数字不可以开头<br>

##  行与缩进
在python中不需要使用大括号来表示代码块的里面，在python中主要使用的是缩进来表示代码块，主要使用tab键（但是不是tab符号）来进行缩进，<br>
一般情况下一个tab缩进是4个空格数，但是缩进的空格数是可变的，然而同一个代码块的语句必须包含相同的缩进空格数

在C语言中是存在块作用域的，块作用域是什么，我们在C语言中已经学过，这里我们主要是学习python，python是不存在块作用域的。</br>
python只存在`函数作用域`,`全局作用域`,`闭包作用域`,`模块作用域`

    L：local，函数（局部）作用域，即函数中定义的变量；
    E：enclosing，嵌套的父级函数的局部作用域（闭包作用域），即包含此函数的上级函数的局部作用域，但不是全局的；
    G：globa，全局变量，就是模块级别定义的变量；
    B：built-in，系统固定模块里面的变量（模块作用域），比如int, bytearray等。 
搜索变量的优先级顺序依次是：作用域局部>外层作用域>当前模块中的全局>python内置作用域，也就是LEGB。
